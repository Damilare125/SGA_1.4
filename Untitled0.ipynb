QUESTION 2:
Using matplotlib module, make a lineplot of 'age' on the x-axis against 'children' on the y-axis using 
the following labels for your plot:
x-axis: 'Age (years)'
y-axis: 'Number of children'
plot title: 'Plot of Age against children'.

```python
data = pd.read_csv("insurance.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# Making a lineplot of Age and Children on the x-axis and y-axis respectively.

plt.figure(figsize=(6, 4))

insurance_csv = data.groupby("age").sum()
var = insurance_csv.reset_index()
x = var["age"]
y = var["children"]
plt.plot(x,y)
plt.xlabel('Age(years)', size = 16)
plt.ylabel('Number of children', size = 16)
plt.title("Plot of age against children", size=20)
plt.show()
```

QUESTION 3:
How would you use matplotlib module to make two different subplots on different axes with the following 
characteristics?

a) First subplot: 
i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p.

b) Second subplot:
i. Create a bar-chart with the 'region' column. 
ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.

```python
# Pie chart using the 'sex" column.
gender = data["sex"].value_counts()
label=["female", "male"]
plt.pie(gender, autopct="%1.2f%%", startangle=90)
plt.legend(label, loc="best")
plt.show()



# A bar-chart with the "region" column.
x_axis = data.groupby("region", as_index = False).sum()
y = data["region"].value_counts()
x = x_axis["region"]
label=["region"]
plt.bar(x,y)
plt.xlabel("Region")
plt.ylabel("Counts")
plt.title("Bar plot for Region")
plt.legend(label, loc="best")
plt.show()
```

QUESTION 4:
Using matplotlib, plot two histograms on the same axis for columns 'age' and 'children'. Ensure the following
is indicated in the plot:
1. The two plots should be clearly visible to the eye [Hint: Use the 'alpha' argument]

 ii. Use appropriate titles for the axes and the plot

```python
# Plotting two histograms on the same axis for Age and Children columns.

plt.figure(figsize=(6,4))

x = (data["age"])
y = (data["children"])
plt.hist(x, alpha=0.5, label="age")
plt.hist(y, alpha=0.5, label="children")
plt.legend(loc="upper right")
plt.title("Histogram of age and children ", size=20)
plt.show()
```

Using the dataset given below,

x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)

y_var = np.random.normal(loc = 50, scale = 1.5, size = 300) 

z_var = np.random.normal(loc = 60, scale = 2, size = 300)

i. Create histograms for the two variables,x_var and y_var, on the same axis. Ensure the two plots can easily 
be see using the appropriate parameter. Also, use the legend() function for the plot.

ii. Make two subplots using the 3 variables as follows:

1. Make a lineplot of y_var against z_var for the first subplot

[Hint: All plots should be well labeled] 

```python
# Creating histograms for two variables; x_var and y_var, on the same axis.
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
plt.hist(x_var, label="x_var")
plt.hist(y_var, label="y_var")
plt.legend(loc="upper left")
plt.title('Histogram of x_var and y_var', size=20)
plt.show()
```

```python
# Making a lineplot of y_var against z_var.
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
z_var = np.random.normal(loc = 60, scale = 2, size = 300)

x = y_var
y = z_var
plt.plot(x,y)
plt.title("Lineplot of y_var against z_var", size=20)
plt.show()



# Making histogram using x_var.
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200)
plt.hist(x_var, bins=10)
plt.title('Histogram of x_var', size=20)
plt.show()
```
